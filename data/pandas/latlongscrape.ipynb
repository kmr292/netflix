```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
from pprint import pprint
import numpy as np
```

```python
# Scrape data from website
res = requests.get("https://developers.google.com/public-data/docs/canonical/states_csv")
soup = BeautifulSoup(res.content,'lxml')
table = soup.find_all('table')[0] 
latlong_table = pd.read_html(str(table))
latlong_table = latlong_table[0]
latlong_table
```

```python
# Round lat and long to 4 decimal places
latlong_table = np.round(latlong_table, decimals=4)
latlong_table
```

```python
# Upload data to csv file
latlong_table.to_csv('latlong.csv')
```
