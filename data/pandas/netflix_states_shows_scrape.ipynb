```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
from pprint import pprint
```

```python
res = requests.get('https://www.highspeedinternet.com/resources/the-most-popular-netflix-show-in-each-state-for-2018')
soup = BeautifulSoup(res.content,'lxml')
table = soup.find_all('table')[0] 
# table = soup.find_all('table')("State"="DC") [Show]
netflix_table = pd.read_html(str(table))
netflix_table = netflix_table[0]
netflix_table

# pprint(df[0].to_json(orient='records'))
```

```python
netflix_table.to_csv('netflix.csv')
```
